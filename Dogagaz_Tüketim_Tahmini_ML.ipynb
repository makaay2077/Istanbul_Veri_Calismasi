```python
"""Bu uygulama da Makine Öğrenimi Algoritmaları ile  İBB Verisetlerinden İlçere Göre Doğalgaz Tüketimi , 

Nüfus ve Sosyal Ekonomik Statü Skoru veri setleri kullanılarak , Sosyoekonmik statü skoru ve Nüfusa göre

Doğalgaz tüketim tahmini yaptık.Bunun için makine öğrenimi kütüphanesi olan sklearn dört ayrı modelini 

kullandık.En iyi metrik doğruluk skorunu veren LinearRegression() modeli idi.Son satırda da tahmini bir 

nüfus ve SES skoru girerek lineer regresyon ile tüketim  tahimini yaptım.
"""


import pandas as pd;

data_ses=pd.read_csv('SES_Skoru.csv' , encoding='UTF-8')


```

```python
data2=data_ses.groupby('İlçe').mean()
data2
```

```python
data1=pd.read_csv('dogalgaz.csv' , sep=';' ,index_col=False)
data1.set_index('id' , inplace=True)
data1
```

```python
df= pd.merge(data1,data2,on='İlçe')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import svm
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,accuracy_score
from sklearn.neighbors import (NeighborhoodComponentsAnalysis,KNeighborsClassifier)


X=df[['Nüfus' , 'SES_SKORU']].values                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                
Y=df['2019'].values.reshape(-1,1)
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.2, random_state=1)


#Random Forest Rassal veya Rastgele Orman
rf=RandomForestRegressor(n_estimators=100 , random_state=42)
rf.fit(x_train,y_train)
y_pred2_rf=rf.predict(x_test)


#Lineear
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred2_reg=reg.predict(x_test)


#Nearest Neighbors Classification En Yakın Komşu Algoritması
knn=KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred2_knn=knn.predict(x_test)



#Support Vector Classification.
clf = svm.SVC()
clf.fit(x_train, y_train)
y_pred2_clf=clf.predict(x_test)


# R-squared score
print('\Rastgele Orman Algoritması R-squared Skoru: ', r2_score(y_test,y_pred2_rf)*100,'\n')
print('\Lineer Yöntem Algoritması R-squared Skoru: ', r2_score(y_test,y_pred2_reg)*100,'\n')
print('\En yakın komşu Algoritması R-squared Skoru: ', r2_score(y_test,y_pred2_knn)*100,'\n')
print('\Destek Vektör Algoritması R-squared Skoru: ', r2_score(y_test,y_pred2_clf)*100,'\n')
```
