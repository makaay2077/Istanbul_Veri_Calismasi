```python
data1=data.drop(columns=["ILCE_UAVT","ADI","ALT_KATEGORI","ADRES","TELEFON","WEBSITESI",
                    "ACIL_SERVIS","AMBULANS","MAHALLE","ENLEM","BOYLAM"])
#data1=data1.groupby(["ILCE_ADI","ENLEM","BOYLAM"]).sum()
data1

data2=pd.read_csv('Nufus.csv',sep=';')
```

```python
data1=data1.groupby(["ILCE_ADI"])
grouped_df = data1.sum()

data10=grouped_df.reset_index()
data10

```

```python

data_20=pd.merge(data10,data2,on=["ILCE_ADI"])

data_20
```

```python
data_20["Nüfus_Yatak_Oranı"]=1000/((data_20["Nüfus"])/data_20["YATAK"])
data_20


import plotly.graph_objects as go

fig = go.Figure(go.Bar(
            x=data_20['Nüfus_Yatak_Oranı'],
            y=data_20["ILCE_ADI"],
            orientation='h'))

fig.show()
```
